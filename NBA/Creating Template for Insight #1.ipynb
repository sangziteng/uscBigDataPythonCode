```python
## Preamble ##

import os 
import pandas as pd
import numpy as np 
```

```python
# Change to File Directory #
```

```python
# Go to Directory with Results #

files =  os.listdir("./TEST 2 ORIGINAL")    
```

```python
# Find Box Scores # 

for file in files[0:2]:
    
    data = pd.read_csv("./TEST 2 ORIGINAL/" + file)
    
    HomeScore = data['home_score'].max()
    AwayScore = data['away_score'].max()

    Year, Month, Day = file.replace("[", "").replace("]", "").split("-")[0:3]

    Matchup = file.replace("[", "").replace("]", "").replace(".csv", "").split("-")[4]

    Away, Home = Matchup.split("@")
    
```

```python
# Set up Data for Plus Minus # 

data2 = data.loc[data['points'].notnull()]
data2 = data2.loc[data['points']> 0 ]
datalist = data2[['a1','a2','a3','a4','a5','h1','h2','h3','h4','h5', 'points','player']].values.tolist()
```

```python
HomePlayers = []
AwayPlayers = []

for i in range(1,6,1):
    
    HomePlayers = HomePlayers + list(data2['h' + str(i)])
    AwayPlayers = AwayPlayers + list(data2['a' + str(i)])
    
HomePlayers = list(set(HomePlayers))
AwayPlayers = list(set(AwayPlayers))
```

```python
HomePlayerList = []
for player in HomePlayers: 

    PlusMinus = 0
    
    for play in datalist: 
        
        HomePlayersOnCourt = play[5:10]
        AwayPlayersOnCourt = play[0:5]

        Score = play[10]
        ScoringPlayer = play[11]
        
        if player in HomePlayersOnCourt:
        
            if ScoringPlayer in HomePlayersOnCourt:

                PlusMinus = PlusMinus + Score

            else:

                PlusMinus = PlusMinus - Score

                
    HomePlayerList.append([player, PlusMinus])
        
HomePlayerList = pd.DataFrame(HomePlayerList , columns = ['Player', 'PlusMinus'])   

HomePlayerList['Team'] = Home
```

```python
AwayPlayerList = []
for player in AwayPlayers: 

    PlusMinus = 0
    
    for play in datalist: 
        
        HomePlayersOnCourt = play[5:10]
        AwayPlayersOnCourt = play[0:5]

        Score = play[10]
        ScoringPlayer = play[11]
        
        if player in AwayPlayersOnCourt:
        
            if ScoringPlayer in AwayPlayersOnCourt:

                PlusMinus = PlusMinus + Score

            else:

                PlusMinus = PlusMinus - Score

                
    AwayPlayerList.append([player, PlusMinus])
        
AwayPlayerList = pd.DataFrame(AwayPlayerList , columns = ['Player', 'PlusMinus'])        
AwayPlayerList['Team'] = Away
```

```python
SortedPlusMinusData = pd.concat([HomePlayerList, AwayPlayerList]).sort_values('PlusMinus', ascending = False)
```

```python
# Putting it All Together 

FinalData = []

for file in files:
    
    data = pd.read_csv("./TEST 2 ORIGINAL/" + file)
    
    HomeScore = data['home_score'].max()
    AwayScore = data['away_score'].max()

    Year, Month, Day = file.replace("[", "").replace("]", "").split("-")[0:3]

    Date = str(Year) + "-" + str(Month) + "-" + str(Day)
    
    Matchup = file.replace("[", "").replace("]", "").replace(".csv", "").split("-")[4]

    Away, Home = Matchup.split("@")
    
    # Set up Data for Plus Minus # 

    data2 = data.loc[data['points'].notnull()]
    data2 = data2.loc[data['points']> 0 ]
    datalist = data2[['a1','a2','a3','a4','a5','h1','h2','h3','h4','h5', 'points','player']].values.tolist()

    # Get List of Home / Away Players #
    
    HomePlayers = []
    AwayPlayers = []

    for i in range(1,6,1):

        HomePlayers = HomePlayers + list(data2['h' + str(i)])
        AwayPlayers = AwayPlayers + list(data2['a' + str(i)])

    HomePlayers = list(set(HomePlayers))
    AwayPlayers = list(set(AwayPlayers))
    
    # Create + / - for Home Players # 

    HomePlayerList = []
    for player in HomePlayers: 

        PlusMinus = 0

        for play in datalist: 

            HomePlayersOnCourt = play[5:10]
            AwayPlayersOnCourt = play[0:5]

            Score = play[10]
            ScoringPlayer = play[11]

            if player in HomePlayersOnCourt:

                if ScoringPlayer in HomePlayersOnCourt:

                    PlusMinus = PlusMinus + Score

                else:

                    PlusMinus = PlusMinus - Score


        HomePlayerList.append([player, PlusMinus])

    HomePlayerList = pd.DataFrame(HomePlayerList , columns = ['Player', 'PlusMinus'])   

    HomePlayerList['Team'] = Home
    
    # Create + / - for Away Players # 

    AwayPlayerList = []
    for player in AwayPlayers: 

        PlusMinus = 0

        for play in datalist: 

            HomePlayersOnCourt = play[5:10]
            AwayPlayersOnCourt = play[0:5]

            Score = play[10]
            ScoringPlayer = play[11]

            if player in AwayPlayersOnCourt:

                if ScoringPlayer in AwayPlayersOnCourt:

                    PlusMinus = PlusMinus + Score

                else:

                    PlusMinus = PlusMinus - Score


        AwayPlayerList.append([player, PlusMinus])

    AwayPlayerList = pd.DataFrame(AwayPlayerList , columns = ['Player', 'PlusMinus'])        
    AwayPlayerList['Team'] = Away

    # Put it all together #
    SortedPlusMinusData = pd.concat([HomePlayerList, AwayPlayerList]).sort_values('PlusMinus', ascending = False)
    
    FinalData.append([file, Date, Matchup, Home, Away, HomeScore, AwayScore] + SortedPlusMinusData.values.tolist()[0] )
    
FinalData = pd.DataFrame(FinalData, columns = ['File','Date', 'Matchup', "Home", "Away", "HScore", "AScore", "Highest +/-", "+/- (Optional)", "Team (Optional)"])
```

```python
FinalData.to_csv("./Insight #1 Answers.csv")

# Now, blacking out the answers for you #

FinalData['HScore'] = np.NaN
FinalData['AScore'] = np.NaN
FinalData['Highest +/-'] = ""
FinalData["+/- (Optional)"] = np.NaN
FinalData["Team (Optional)"] = ""
```
