## Demonstration of Play-by-Play Data Analysis

Say that, we want to understand something about the + / -, but now at the level of the individual players. 

I will again focus on the Lakers (LAL) and compare the performance of those players in individual matchups. 

```python
# Load Libraries # 

import os
import pandas as pd 
import numpy as np 

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
## Create a list of all Lakers game files. (ie. files where the naem contains LAL) #

# Here, my path is : "C:/Users/mmiric/Documents/NBA Data/1718 NBA Data/TRAIN", but yours 
# might differ #

LAL_Files = []

for file in os.listdir("1718/TRAIN"):
    if "LAL" in file: 
        LAL_Files.append(file)
```

```python
# Let's see the head (first 5 rows) in the first data file: 

Directory = "1718/TRAIN/"

data = pd.read_csv(Directory + LAL_Files[0])

data.head()
```

```python
# Now, we have the lineups separate for the home and away team, but we dont have an indicator in the dataset
# for what is the home and away team. Therefore, we have to create separate datasets for 
# home / away teams. 

# Let's build the code for a single game, then we run for all games #. 


Directory = "1718/TRAIN/"
file =  LAL_Files[0]

data = pd.read_csv(Directory + file)

# DEFINE IF HOME OR AWAY GAME # 

if "@LAL" in file: 
    HomeAway = "Home"
else: 
    HomeAway = "Away"

# CLEAN HOME DATA # 

if HomeAway == "Home":
    
    # Select only key columns # 
    
    temp_data = data[['h1', 'h2', 'h3', 'h4', 'h5','a1', 'a2', 'a3', 'a4', 'a5','player', 'team', 'points']]
    
    temp_data = temp_data.loc[temp_data['points'].notnull()]
    
    # Now, we are going to focus on a "player - position pair", and how many players scored / conceded points. 
    
    i = str(1)
    
    # Select only player and opponent #
    
    temp_data2 = temp_data[['h'+i, 'a'+i, 'player', 'team', 'points']]
    
    temp_data2.columns = ['LALPlayer', "OpponentPlayer", "Scorer", "Team", "Points"]
    
    # Construct Variables indicating whether a player has scored, or not.# 
    
    temp_data2['Team'] = np.where(temp_data2['Team'] == "LAL", 1, 0)
    
    temp_data2['Player'] = np.where(temp_data2['LALPlayer'] == temp_data2['Scorer'], 1, 0)
    
    temp_data2['OpposingPlayer'] = np.where(temp_data2['OpponentPlayer'] == temp_data2['Scorer'], 1, 0)

    temp_data2['ScoredMissed'] = np.where(temp_data2['Points'] == 0, "Missed", "Scored")
    
    temp_data2['Position'] = i
    
    # Calculate 
    
    temp_data2 = pd.pivot_table(temp_data2, index = ['LALPlayer', 'ScoredMissed'], values = ['Team','Player','OpposingPlayer'], aggfunc = np.sum)
    
   # print(temp_data2)
    print(temp_data2)
    
  #  temp_data2 = temp_data2.unstack()
    
  #  print(to_numerci(temp_data2[[('Team', 'Scored')]]- temp_data2[[('Team', 'Missed')]])

#temp_data2[[('Team', 'Scored')]].stack()


```

```python
# temp_data2.loc[('Brandon Ingram', 'Scored')].loc['Team'] - temp_data2.loc[('Brandon Ingram', 'Missed')].loc['Team']

# Calculate Player's PlusMinus
# Create a dataframe 'temp_data3' to accept and record each play's +/- in each Game
temp_data3 = temp_data2.unstack(level=1)['Team']['Scored'] - temp_data2.unstack(level=1)['Team']['Missed']
temp_data3 = temp_data3.reset_index()
temp_data3 = temp_data3.rename(columns={'LALPlayer':'Player', 0:'PlusMinus'})
temp_data3['Team'] = 'LAL'
temp_data3['GameID'] = data['game_id']
temp_data3['HomeAway'] = HomeAway

# attach current game data to playerPlusMinus total dataset to accept and record each play's +/- in each Game
playerPlusMinus = pd.DataFrame()
playerPlusMinus = playerPlusMinus.append(temp_data3, ignore_index=True)

playerPlusMinus.head()
```

```python
# If we add this to our original code, and expand the code for Away Games, we can get the following code #

# Now, we have the lineups separate for the home and away team, but we dont have an indicator in the dataset
# for what is the home and away team. Therefore, we have to create separate datasets for 
# home / away teams. 

# Let's build the code for a single game, then we run for all games #. 

Directory = "1718/TRAIN/"

# Create an Empty Data Frame so that we can merge # 

FinalData = pd.DataFrame(columns = ['Lineup', 'ScoredConceded', 'points'])

for file in LAL_Files:
    
    data = pd.read_csv(Directory + file)

    if "@LAL" in file: 
        HomeAway = "Home"

        temp_data = data[['h1', 'h2', 'h3', 'h4', 'h5', 'team', 'points']]

        temp_data.columns = ['p1', 'p2', 'p3', 'p4', 'p5', 'team', 'points']

        temp_data['ScoredConceded'] = np.where(temp_data['team'] == "LAL", "Scored", "Conceded")    

        temp_data = temp_data.loc[temp_data['points'].notnull()]

        names = temp_data[['p1', 'p2', 'p3', 'p4', 'p5']].values.tolist()

        SortedLines = []

        for line in names: 

            SortedLines.append( sorted(line) )

        SortedLines = pd.DataFrame(SortedLines, columns = ['p1', 'p2', 'p3', 'p4', 'p5'])

        SortedLines['Lineup'] = SortedLines[['p1', 'p2', 'p3', 'p4', 'p5']].apply(lambda x: ', '.join(x), axis=1)

        temp_data['Lineup'] = SortedLines['Lineup']
        
        temp_data = temp_data[['Lineup', 'ScoredConceded', 'points']]

        # Combine Data Files #

        FinalData = pd.concat([FinalData, temp_data])

    else:
        HomeAway = "Away"

        temp_data = data[['a1', 'a2', 'a3', 'a4', 'a5', 'team', 'points']]

        temp_data.columns = ['p1', 'p2', 'p3', 'p4', 'p5', 'team', 'points']

        temp_data['ScoredConceded'] = np.where(temp_data['team'] == "LAL", "Scored", "Conceded")    

        temp_data = temp_data.loc[temp_data['points'].notnull()]

        names = temp_data[['p1', 'p2', 'p3', 'p4', 'p5']].values.tolist()

        SortedLines = []

        for line in names: 

            SortedLines.append( sorted(line) )

        SortedLines = pd.DataFrame(SortedLines, columns = ['p1', 'p2', 'p3', 'p4', 'p5'])

        SortedLines['Lineup'] = SortedLines[['p1', 'p2', 'p3', 'p4', 'p5']].apply(lambda x: ', '.join(x), axis=1)

        temp_data['Lineup'] = SortedLines['Lineup']

        temp_data = temp_data[['Lineup', 'ScoredConceded', 'points']]

        FinalData = pd.concat([FinalData, temp_data])

```

```python
FinalData
```

```python
# Collapse the data (pivot) using a pivot table to calculate the number of points scored / conceded #

FinalData2 = pd.pivot_table(data = FinalData, index = ['Lineup','ScoredConceded'], values = 'points', aggfunc = [np.sum])

FinalData2 = FinalData2.unstack().reset_index()

# This command just resets the header, so that it is simpler to work with the column names#

FinalData2.columns = FinalData2.columns.get_level_values(0)

# This defines new column names #

FinalData2.columns = ['Lineup', 'Conceded', 'Scored']

# This calculates the +/-

FinalData2['PlusMinus'] = FinalData2['Scored'] - FinalData2['Conceded']

FinalData2
```

```python
# Sort the Pivot Table as Desired #

FinalData2.sort_values('PlusMinus', ascending = True)
```
